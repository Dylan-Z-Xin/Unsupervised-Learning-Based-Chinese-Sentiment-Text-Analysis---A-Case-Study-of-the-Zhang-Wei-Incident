```python
#ç¬¬ä¸€æ­¥ï¼šæ„é€ ç‰¹å¾ç©ºé—´å’Œæ ‡ç­¾

Y=[]
for ind in review_long_clean.index_content.unique():
    y=[ word for word in review_long_clean.content_type[review_long_clean.index_content==ind].unique() ]
    Y.append(y)
len(Y)

X=[]
for ind in review_long_clean.index_content.unique():
    term=[ word for word in review_long_clean.word[review_long_clean.index_content==ind].values ]
    X.append(' '.join(term))
len(X)

X
Y
```

```python
#ç¬¬äºŒæ­¥ï¼šè®­ç»ƒé›†ã€æµ‹è¯•é›†åˆ’åˆ†
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=7)

#ç¬¬ä¸‰æ­¥ï¼šè¯è½¬å‘é‡ï¼Œ01çŸ©é˜µ
count_vec=CountVectorizer(binary=True) 
x_train=count_vec.fit_transform(x_train)
x_test=count_vec.transform(x_test)  

#ç¬¬å››æ­¥ï¼šæ„å»ºå†³ç­–æ ‘
dtc=tree.DecisionTreeClassifier(max_depth=5)
dtc.fit(x_train,y_train)
print('åœ¨è®­ç»ƒé›†ä¸Šçš„å‡†ç¡®ç‡ï¼š%.2f'% accuracy_score(y_train,dtc.predict(x_train)))

y_true=y_test
y_pred=dtc.predict(x_test)
print(classification_report(y_true,y_pred))
print('åœ¨æµ‹è¯•é›†ä¸Šçš„å‡†ç¡®ç‡ï¼š%.2f'% accuracy_score(y_true,y_pred))
```

# åŸºäºLDAæ¨¡å‹çš„ä¸»é¢˜åˆ†æ

```python
data=review_mltype.copy()

word_data_pos=data[data['ml_type']=='pos']
word_data_neg=data[data['ml_type']=='neg']
```

```python
#å»ºç«‹è¯å…¸ï¼Œå»é‡

pos_dict=corpora.Dictionary([ [i] for i in word_data_pos.word]) #shape=(n,1)
neg_dict=corpora.Dictionary([ [i] for i in word_data_neg.word])
```

```python
print(pos_dict)
```

```python
#å»ºç«‹è¯­æ–™åº“

pos_corpus=[ pos_dict.doc2bow(j) for j in [ [i] for i in word_data_pos.word] ] #shape=(n,(2,1))
neg_corpus=[ neg_dict.doc2bow(j) for j in [ [i] for i in word_data_neg.word] ]
```

```python
len(word_data_pos.word)
len(pos_dict)
len(pos_corpus)
pos_corpus
```

```python
#æ„é€ ä¸»é¢˜æ•°å¯»ä¼˜å‡½æ•°

def cos(vector1,vector2):
    '''
    å‡½æ•°åŠŸèƒ½ï¼šä½™ç„ç›¸ä¼¼åº¦å‡½æ•°
    '''
    dot_product=0.0
    normA=0.0
    normB=0.0
    for a,b in zip(vector1,vector2):
        dot_product +=a*b
        normA +=a**2
        normB +=b**2
    if normA==0.0 or normB==0.0:
        return None
    else:
        return ( dot_product/((normA*normB)**0.5) )
```

```python
#ä¸»é¢˜æ•°å¯»ä¼˜
#è¿™ä¸ªå‡½æ•°å¯ä»¥é‡å¤è°ƒç”¨ï¼Œè§£å†³å…¶ä»–é¡¹ç›®çš„é—®é¢˜

def LDA_k(x_corpus,x_dict):
    '''
    å‡½æ•°åŠŸèƒ½ï¼š
    '''
    #åˆå§‹åŒ–å¹³å‡ä½™ç„ç›¸ä¼¼åº¦
    mean_similarity=[]
    mean_similarity.append(1)
    
    #å¾ªç¯ç”Ÿæˆä¸»é¢˜å¹¶è®¡ç®—ä¸»é¢˜é—´ç›¸ä¼¼åº¦
    for i in np.arange(2,11):
        lda=models.LdaModel(x_corpus,num_topics=i,id2word=x_dict) #LDAæ¨¡å‹è®­ç»ƒ
        
        for j in np.arange(i):
            term=lda.show_topics(num_words=50)
        
        #æå–å„ä¸»é¢˜è¯
        top_word=[] #shape=(i,50)
        for k in np.arange(i):
            top_word.append( [''.join(re.findall('"(.*)"',i)) for i in term[k][1].split('+')]) #åˆ—å‡ºæ‰€æœ‰è¯
        
        #æ„é€ è¯é¢‘å‘é‡
        word=sum(top_word,[]) #åˆ—è½¦æ‰€æœ‰è¯
        unique_word=set(word) #å»é‡
        
        #æ„é€ ä¸»é¢˜è¯åˆ—è¡¨ï¼Œè¡Œè¡¨ç¤ºä¸»é¢˜å·ï¼Œåˆ—è¡¨ç¤ºå„ä¸»é¢˜è¯
        mat=[] #shape=(i,len(unique_word))
        for j in np.arange(i):
            top_w=top_word[j]
            mat.append( tuple([ top_w.count(k) for k in unique_word ])) #ç»Ÿè®¡listä¸­å…ƒç´ çš„é¢‘æ¬¡ï¼Œè¿”å›å…ƒç»„
        
        #ä¸¤ä¸¤ç»„åˆã€‚æ–¹æ³•ä¸€
        p=list(itertools.permutations(list(np.arange(i)),2)) #è¿”å›å¯è¿­ä»£å¯¹è±¡çš„æ‰€æœ‰æ•°å­¦å…¨æ’åˆ—æ–¹å¼ã€‚
        y=len(p) # y=i*(i-1)
        top_similarity=[0]
        for w in np.arange(y):
            vector1=mat[p[w][0]]
            vector2=mat[p[w][1]]
            top_similarity.append(cos(vector1,vector2))

#        #ä¸¤ä¸¤ç»„åˆï¼Œæ–¹æ³•äºŒ
#        for x in range(i-1):
#            for y in range(x,i):
            
        #è®¡ç®—å¹³å‡ä½™ç„ç›¸ä¼¼åº¦
        mean_similarity.append(sum(top_similarity)/ y)
    return mean_similarity
```

```python
#è®¡ç®—ä¸»é¢˜å¹³å‡ä½™ç„ç›¸ä¼¼åº¦

pos_k=LDA_k(pos_corpus,pos_dict)
neg_k=LDA_k(neg_corpus,neg_dict)

pos_k
neg_k
```
